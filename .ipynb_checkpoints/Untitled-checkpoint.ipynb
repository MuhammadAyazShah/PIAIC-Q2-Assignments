```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
rootDir = 'pictures'
for dirName, subdirList, fileList in os.walk(rootDir):
    print('Found directory: %s' % dirName)
    for fname in fileList:
        print('\t%s' % fname)
```

```python
origin_list = []
pic = 1
fig=plt.figure(figsize=(15,15))
columns=4
rows=5
for i in range(1,len(fileList)+1):
    im = Image.open(dirName+"\\"+fileList[i-1])
    fig.add_subplot(rows,columns,i)
    plt.imshow(im)
    print("\nPicture #:", pic)
    original_array = np.array(im)
    print("Shape of Picture #" , pic, ":", original_array.shape, "\nSize of Picture #" , pic, ":", original_array.size, "\nDimension of Picture #" , pic, ":", original_array.ndim, "\nArray of Picture #" , pic, ":", original_array)
    print("\n************")
    origin_list.append(im)

plt.show()
print("APPENDING ORIGINAL PICTURES")
print(len(origin_list))
```

```python
main_list = []
pic = 1
fig=plt.figure(figsize=(11,12))
columns=4
rows=5
for i in range(1,len(fileList)+1):
    print("\nResize Picture #:", pic)
    new_image = origin_list[i-1].resize((200, 200))
    resized_array = np.array(new_image)
    fig.add_subplot(rows,columns,i)
    plt.imshow(resized_array)
    print("Shape of Resized Picture #" , pic, ":", resized_array.shape, "\nSize of Resized Picture #" , pic, ":", resized_array.size, "\nDimension of Resized Picture #" , pic, ":", resized_array.ndim, "\nArray of Resized Picture #" , pic, ":", resized_array)
    print("\n************")
    
    main_list.append(resized_array)
    pic = pic+1
    #print(plt.imshow(main_list[i]))
plt.show()
print("\nAPPENDING RESIZED PICTURES:-")
print(len(main_list))
```

```python
def main():
    main_arr= np.array(main_list)
    print("FINAL NUMPY ARRAY :\n\n ",main_arr)
    return main_arr

main_arr = main()
print(main_arr.shape)
print(main_arr.size)
print(main_arr.ndim)
```
